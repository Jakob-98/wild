```python
import cv2
import os
import glob
from PIL import Image
from pathlib import Path
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import mahotas
from tqdm import tqdm
```

```python
ena_local = 'C:/temp/ena/images/test/'
images = [os.path.split(i)[1] for i in glob.glob(ena_local + '/*.jpg', recursive=True)]
```

```python
from collections import OrderedDict
import multiprocessing as mp
```

```python
from tqdm.contrib.concurrent import process_map  # or thread_map
import time

feats = {}
def _getfeat(image):
    img = cv2.imread(ena_local + imx)
    b, g, r =  (mahotas.features.lbp(img[:, :, i], 2.5, 12) for i in (0, 1, 2))
    lpbrgb = np.stack((b,g,r))
    chists_reshaped = np.squeeze(np.stack(tuple(cv2.calcHist(img, [i], None, [352], [0,256]) for i in (0, 1, 2))))
    vstack = np.vstack((lpbrgb, chists_reshaped))
    feats[img] = vstack


process_map(_getfeat, images, max_workers=6, chunksize=2)
```

```python
fimages = np.ndarray((len(images), 6, 352))
for i, imx in enumerate(tqdm(images)):
    img = cv2.imread(ena_local + imx)
    b, g, r =  (mahotas.features.lbp(img[:, :, i], 2.5, 12) for i in (0, 1, 2))
    lpbrgb = np.stack((b,g,r))
    chists_reshaped = np.squeeze(np.stack(tuple(cv2.calcHist(img, [i], None, [352], [0,256]) for i in (0, 1, 2))))
    vstack = np.vstack((lpbrgb, chists_reshaped))
    fimages[i] = vstack
```
